```python
#문자열은 어떻게 만들고 사용할까?

##1. 큰따옴표로 양쪽 둘러싸기
a = "hello world"

##2. 작은 따옴표로 양쪽 둘러싸기
b = 'python is fun'

##3. 큰따옴표 3개로 둘러싸기
c = """Life is too short, You need python"""

##4. 작은 따옴표 3개로 뚤러싸기
d = '''Life is too short, You need python'''

print(a)
print(b)
print(c)
print(d)

print(type(a)) # 컨트로 쉬프트 알트 
print(type(b))
print(type(c))
print(type(d))
```

```python
# 문자열안에 작은 따옴표 포함하기
a = "Python's favorite food is perl"
c = 'Python\'s favorite food is perl' #에러임 작은따옴표 써서 ,역슬레쉬 쓸것.
b = """Python's favorite food is perl"""
print(c)
```

```python
# 문자열 안에 큰 따옴표 포함하기

b = "\"Python is very easy\" he says." # \뒤에있는거는 그대로 표현한다는 의미
a = 'Python is very easy" he says.'
print(b)
```

```python
#문자열 포매팅

## 1. 숫자 바로 대입
print("i eat %d apples" %3) #숫자일때는 % 뒤에 d

## 2. 문자열 바로 대입
print("i eat %s apple" %"five") #문자일때는 % 뒤에 s
```

```python
## 3. 숫자값을 나타내는 변수로 대입
number = 3
print("i eat %d apples" % number)
```

```python
## 2개 이상의 값 넣기
number = 10
day = "three"
print("i eat %d apples so i was sick for %s days" % (number, day)) #순서에 맞게 변수 적기
```

```python
# format 함수를 사용한 포매팅 %s, %d 쓸필요없이 쓸수있음.

print("i eat {0} apples".format(3))
print("i eat {0} apples".format("five"))
```

```python
# format 함수를 이용해서 두개이상의 값 넣기
number = 10
day = "three"
print("i eat {0} apples so i was sick for {1} days".format(number, day)) #중괄호 0 ,1 순서는 format(number , day) 순서와 동일
```

```python
# f문자열 포매팅

name = "홍길동"
age = 10
print(f"나의 이름은 {name} 입니다. 나이는 {age} 입니다.") #큰따옴표 앞에 f 적기(편한방법임) 변수는 중괄호로 쓸것.
```

```python
#문자열에 관련된 함수들

a = "hobby" #a는 문자열에 성격을 가진 객체이다 라는 뜻.
print(type(a))

## 문자의 갯수 세기 - count
print(a.count("b")) # b의 글자가 몇개인지 세아림 .count
```

```python
## 위치 알려주기 - find

a = "Python is the best choice"
print(a.find("b")) #b의 위치 알려줌.
print(a.find("k")) #없으면 -1 값을 반환함.
```

```python
# 위치 알려주기2 - index 
a = "Life is too short"
print(a.index("t")) #첫번쨰 t위치만 알려줌
print(a.index("k")) #에러가 뜸.
```

```python
# 문자열 삽입 - join

print(",".join("abcd")) #abcd 사이에 , 삽입하는 것.
```

```python
# 소문자를 대문자로 바꾸기 - upper
a = "hi"
a.upper()
```

```python
# 대문자를 소무자로 바꾸기 - lower
a = "HI"
a.lower()
```

```python
# 공백지우기 - strip
a = "   hi   "
print(a)
a = a.strip()
print(a)
```

```python
# 공백지우기 왼쪽 / 오른쪽
a = "   hi   "
print(a.lstrip()) #왼쪽 공백은 엘
print(a.rstrip()) #우측 공백은 알
```

```python
# 문자열 바꾸기 - replace
a = "Life is too short"
print(a.replace("Life","your leg"))
```

```python
# 문자열 나누기 - split 
a = "Life is too short"
print(a.split()) #공백을 기준으로 분리를 함. 분홍색 괄호안에 아무것도 없을때.
```
