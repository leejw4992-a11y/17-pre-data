```python
# 튜플은 어떻게 만들까? (tuple)
t1 = ()
t2 = (1, ) # , 반드시 붙일것.
t3 = (1, 2, 3)
t4 = 1, 2, 3
t5 = ('a', 'b', ('ab', 'cd'))

print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))
print(t1)
print(t2)
print(t3)
print(t4)
print(t5)
```
