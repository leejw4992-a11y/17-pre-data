```python
test_list = ['one', 'two', 'three']

for tl in test_list:
	print(tl)
```

```python
a = [(1,2), (3,4),(4,6)] #3개의 요소 / 각각의 요소는 튜플형태

for (k,j) in a : 
	print(j)  # value 값만 출력 
```

```python
marks = [90, 25, 67, 45, 80]

number = 0 #초기화시 필요한 값 임의의 변수로 표현
for mark in marks:
	number = number + 1
	if mark >=60:
		print(f"{number} 학생은 합격입니다.")
	else:
		print(f"{number} 학생은 불합격 입니다.")
```

```python
# for문과 continue

marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
	number +=1
	if mark <=60:
		continue

	print(f"{number} 학생은 합격입니다.")
```

```python
# for와 range를 이용한 구구단
for i in range(2,4):
	for j in range(1,9+1):
		print(f"{i}x{j}= {i*j}")
```

```python
# 리스트 컴프리헨션 사용하기

a = [1,2,3,4]
result = []
for num in a:
	result.append(num*3)

print(result)
```

```python
a = [1,2,3,4]
result = [num*3 for num in a] #앞의 num은 임의의 값이 아님
print(result)  
```

```python
# for문과 break문

for i in range(10) : # range(0,10)
	if i == 5:
		break
	print(i)
```

```python
# enumerate() 함수  # 서비스로 인덱스 번호를 주는 함수.

fruits = ['apple', 'banana', 'orange']

for (i,fruit) in enumerate(fruits):
	print(f"{i+1}번째 과일 : {fruit}")
```

```python
# zip 함수로 여러 리스트 순회하기 / 튜플형태로 출력됨.
names = ['홍길동', '김철수', '이영희']
scores = [85,92,78]

for name, score in zip(names, scores):
	print(f"{name}: {score}점")
```
