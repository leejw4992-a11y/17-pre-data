```python
# 딕셔너리는 어떻게 만들까? {} 쓰기

# {key1: value1, key2: value2, ...} #키와 밸류가 하나의 쌍으로 이루어진것이 딕셔너리임.
#  이름    홍길동   나이   23살         이름은 홍길동 나이는 23살 이런 뜻

dic = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(dic)
print(type(dic))
```

```python
a=1
a=2
print(a) # 변경이 아니라 재할당된 의미임 
```

```python
# 딕셔너리의 key값으로 올 수 없는 데이터 타입은 리스트다. -> mutable

a=[1,2,3]
a[0] = 3
print(a) #3,2,3으로 바뀜.
```

```python
# 딕셔너리 쌍 추가하기

a = {1 : "a"}
a[2] = 'b' # a[2]에서 2는 인덱스 번호가 아닌 딕셔너리의 key값임.

print(a)
```

```python
a[3] = [1,2,3]
print(a)  #결과값은 , 기준으로 추가됨
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a) # 1:'a' 요소 값만 지워짐 (하위 결과값)
```

```python
# 딕셔너리에 key를 사용해서 value 얻기

grade = {
	'pey' : 10,
	'juli' : 99
}

print(grade['pey']) #key 값은 'pey'
print(grade['juli']) #key 값은 'juli'
```

```python
# 딕셔너리 관련 함수
## 1. key 리스트 만들기 - keys
a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
a.keys() # key값을 가져오는것.
list(a.keys()) # 위에것을 토대로 리스트 만들기
```

```python
## value 리스트 만들기 - values
a.values() #value 값 가져오기
list(a.values()) #리스트 만들기
```

```python
## key, value 쌍으로 얻기 - items
print(a)

a.items()
list(a.items()) # 각각의 요소는 튜플로 되어있음 총 3개의 요소임.
```

```python
# key와 value 쌍 모두 지우기 - clear

a.clear() #안에 있는 내용 지우기
print(a)
```

```python
# key로 value 얻기 = get

a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(a['name'])  #이전에 배웠던 것.
print(a.get('name'))
```

```python
# print(a['hobby']) #없는값 출력시 키에러 뜸.
a.get('hobby',0) #에러 안뜨고 없는값 찾는방법
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in

print(a)
print('name' in a)
print('email' in a)
```

```python
# key로 value 꺼내기 - pop

a = {
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}

result = a.pop("phone") #원하는 것만 뽑아내기
print(result) # 전화번호 벨류만 빼냄
print(a) # 전화번호 벨류 빠진 후 남은값 표시
```
