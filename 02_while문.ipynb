```python
treehit = 0
while treehit < 10:
	treehit = treehit + 1 #반복할때마다 트리힛에 +1
	print(f"나무를 {treehit}번 찍었습니다")
	if treehit == 10:
		print('나무가 넘어갑니다')
```

```python
accident = 0
while accident <4:
	accident = accident +1
	print(f"차를 {accident}번 박았습니다")
	if accident == 4:
		print('보험이 말소됬습니다')
```

```python
coffee = 5
money = 300

while money:
	print('돈을 받았으니 커피를 줍니다')
	coffee = coffee - 1
	print(f"남은 커피는 {coffee}개 입니다")

	if coffee == 0:
		print("커피가 다 떨어졌습니다")
		break #멈추는 명령어
```

```python
#자판기 로직
coffee = 5
while True:
	money = int(input('돈을 넣어주세요:'))
	if money == 300:
		print("커피를 줍니다")
		coffee -= 1
	elif money >300:
		print(f"거스름돈 {money -300}를 주고 커피를 줍니다")
		coffee -=1
	else:
		print('돈을 다시 돌려주고 커피를 주지 않습니다')
		print(f"현재 남은 커피는 {coffee}개 입니다")

	if coffee == 0:
		print("커피가 다 떨어졌습니다")
		break
```

```python
# while 문의 맨 처음으로 돌아가기

a = 0
while a < 10:
	a = a + 1
	if a % 2 == 0:
		continue #조건을 충족해서 만나면 while 문으로 맨 처음 돌아감.
	print(a)
```
