```python
# 리스트는 어떻게 만들고 사용할까?

# 리스트명 = [요소1, 요소2, ...]

a = []
b = [1, 2, 3]
c = ["life", "is", "too", "short"]
d = [1, 2, "life", "is"] #리스트 안에 복합도 가능함.
e = [1, 2, ["life", "is"]] #리스트 안에 리스트도 가지고있을수 있음.

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
# 리스트의 인덱싱과 슬라이싱 (중요,데이터분석시 사용)

a = [1, 2, 3]
print(a[0])
print(a[2])
print(type(a[2])) #결과값은 숫자임. 리스트 아님.
```

```python
print(a[0] + a[2])
print(type(a[0] + a[2]))
```

```python
print(a[-1])
```

```python
b = ['1', '2', '3', 'a', 'b', 'c']
print(b[-1])
```

```python
b[0]+b[2]+b[-1] #문자로 취급되서 붙어짐
```

```python
c = [1, 2, ['a', 'b', ['life', 'is']]]
len(c) #요소는 3개임.

c[-1][-1][0]
```

```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]
print(a[0:1])
print(a[0:2])
print(a[0:0])
print(a[3:3])
```

```python
print(a[2:])
print(a[:3])
```

```python
a = [1, 2, 3,['a','b','c'],4,5]

#['b','c']
len(a)

a[3][1:] #'b','c'
a[3][1:2]
```

```python
# 리스트 더하기(+)

a = [1, 2, 3]
b = [4, 5, 6]

print(a+b)
```

```python
# 리스트의 곱하기(*)

a = [1, 2, 3]
a * 3
```

```python
# 리스트의 수정과 삭제

a = [1, 2, 3]
a[2] = 4
print(a)
```

```python
# del 로 리스트의 요소를 삭제하기

a = [1, 2, 3]
del a[1]
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트 관련 함수
## 리스트 요소 추가하기 - append

a = [1, 2, 3]
a.append(4)
print(a)
```

```python
a.append([5, 6])
```

```python
print(a) #원래는 [1, 2, 3, 4, [5, 6]] 으로 결과값이 나옴. / 그전에 미리 추가로 실행시켜서 중복되는 현상임.
```

```python
# 리스트 정렬 - sort
a = [1, 4, 2, 3]
a.sort()
print(a)

a.sort(reverse=True)
print(a)
```

```python
a = [1, 4, 2, 3]
sorted(a)
sorted(a, reverse=True)
```

```python
# 리스트 뒤집기 - reverse
a = ['a', 'c', 'b']

a.reverse()
print(a)
```

```python
# 인덱스 반환 - index / find 함수 참조
a = ['a', 'c', 'b']
a.index("c")
```

```python
# 리스트의 요소 삽입 - insert / 마지막에 계속 집어넣을때는 append 사용
a = [1, 2, 3]
a.insert(0, 4) #처음에 4를 넣어라 라는 의미
print(a)
```

```python
# 리스트의 요소 제거 - remove
a = [1, 2, 3, 1, 2, 3]

a.remove(3)
a.remove(3)
print(a)
```

```python
# 리스트의 요소 꺼내기 - pop
a = [1, 2, 3]
result = a.pop() #빈칸 비워놓으면 -1 인덱스 값 빠짐 / 마지막값
print(a) #결과값은 3이 빠진값이됨.
print(result)
```

```python
# 리스트에 포함된 요소 x의 개수세기 - count

a = [1, 2, 3, 1]
a.count(1)
```

```python
# 리스트 확장 - extend
a = [1, 2, 3]
a.extend([4,5]) #요소로 더해지지 않음 리스트로 한꺼번에 풀려져서 저장됨. / append는 요소로 저장됨.
print(a)
```
