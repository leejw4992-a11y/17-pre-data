```python
# 불 자료형은 어떻게 사용할까?
a = True
b = False 

print(a)
print(b)

print(type(a))
print(type(b))
```

```python
print(1 == 1)
print(2 > 1)
print(2 < 1)
print(2 >= 2)
print(2 <= 2)
print(2 != 2) # 다르다라는 표시
```

```python
# if 조건문(=불자료형이 올 자리임): 
	# 수행할 문장

print(bool("python")) 
print(bool("")) # 이것을 제외하곤 전부 트루임.
print(bool(" "))

print(bool(10))
print(bool(0)) #0빼고는 전부 True라고 보면됨

print(bool([1,2,3]))
print(bool([]))

print(bool({"name": "pey"})) #딕셔너리 활용.
print(bool({}))
```

```python
# 논리연산자 

## and 연산자
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False) # 하나라도 다르면 False 출력
```

```python
## or 연산자

print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True or False)
```

```python
# not 연산자
print(not True) #반대값을 뽑는다 생각하면됨.
```
