```python
# 집합자료형은 어떻게 만들까? -set 함수사용 or {} 쓰기

s1 = set([1,2,3])
s2 = {4,5,6}

print(s1)
print(type(s1))
print(s2)
print(type(s2))
```

```python
s3 = set("hello")
print(s3) #중복 제거 + 순서가 다 다름(랜덤은 아님)
```

```python
# 교집합 구하기 -.intersection()

s1 =set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print(s1)
print(s2)

## .intersection()
print(s1.intersection(s2)) #집합자료형태일때만 사용가능함.

## &
print(s1 & s2)
```

```python
# 합집합 구하기

## .union()
print(s1 .union(s2))

## |
print(s1 | s2)
```

```python
# 차집합 구하기

## difference() -> s1 - s2 한것
print(s1.difference(s2))
print(s1 - s2)

## s2-s1 한것.
print(s2.difference(s1))
print(s2 - s1)
```

```python
# 집합자료형의 관련된 함수

## 값 1개 추가하기 - add
s1 = {1,2,3}
s1.add(4)

print(s1)
```

```python
## 값 여러개 추가하기 - update
s1 = {1,2,3}
s1.update([4,5,6])

print(s1)
```

```python
## 특정값 제거하기 - remove
s1 = {1,2,3}
s1.remove(2)

print(s1)
```

```python
## 모든값의 제거하기 - clear

s1 = {1,2,3}
s1.clear()

print(s1)
```

```python
s1 = set()
s2 = {} #딕셔너리 표시임 집합자료형이 아님

print(type(s1), type(s2))
```
